```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/raw/EdStatsData.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df['Country Code'].nunique()
```

```python
usa = df[df['Country Code'] == 'USA']
usa
```

```python
usa_per_gdp = usa[usa['Indicator Code'] == 'SE.XPD.TOTL.GD.ZS']
usa_per_gdp
```

```python
y = usa_per_gdp.drop(['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], axis=1).dropna(axis=1)
x = y.columns
y = y.values.flatten()
```

```python
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(8, 4), dpi=100)

ax.plot(x, y, color='m', alpha=0.5, label='USA')
ax.set_xlabel('Year')
ax.set_ylabel('% of GDP')
ax.set_xticks([str(x) for x in range(1986, 2015, 4)])
ax.set_title('Education spending, % of US GDP')
plt.tight_layout()
plt.savefig('../images/us_spend_gdp.png')
plt.show()
```

```python
df2 = pd.read_csv('../data/raw/EdStatsData.csv')
```

```python
swe = df2[df2['Country Code'] == 'SWE']
swe
```

```python
swe_per_gdp = swe[swe['Indicator Code'] == 'SE.XPD.TOTL.GD.ZS']
swe_per_gdp
```

```python
b = swe_per_gdp.drop(['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], axis=1).dropna(axis=1)
```

```python
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(8, 4), dpi=100)

ax.plot(a, b, color='m', alpha=0.5, label='Sweden')
ax.set_xlabel('Year')
ax.set_ylabel('% of GDP')
ax.set_xticks(['1979', '1991', '1995', '1999', '2003', '2007', '2011', '2014'])
ax.set_title('Education spending, % of Swedish GDP')
plt.tight_layout()
plt.savefig('../images/swe_spend_gdp.png')
plt.show()
```
