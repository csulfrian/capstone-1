```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams.update({'font.size': 16})
plt.style.use('bmh')
```

```python
df = pd.read_csv('../data/raw/EdStatsData.csv')
```

```python
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

```python
df['Country Code'].nunique()
```

```python
usa = df[df['Country Code'] == 'USA']
usa
```

```python
usa_per_gdp = usa[usa['Indicator Code'] == 'SE.XPD.TOTL.GD.ZS']
usa_per_gdp
```

```python
y = usa_per_gdp.drop(['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], axis=1).dropna(axis=1)
years = list(y[::-1])
y = y.values.flatten()
```

```python
fig, ax = plt.subplots(figsize=(12, 8), dpi=300)

ax.plot(y, color='m', alpha=0.5, label='USA')
ax.set_xlabel('Year')
ax.set_ylabel('% of GDP')
ax.set_xticklabels(years, rotation=80)
ax.set_title('Education spending, % of US GDP')
plt.tight_layout()
plt.savefig('../images/us-spend-gdp.png')
plt.show()
```
