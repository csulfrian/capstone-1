```python
usa = df[df['Country Code'] == 'USA']
usa
```

```python
usa_per_gdp = usa[usa['Indicator Code'] == 'SE.XPD.TOTL.GD.ZS']
usa_per_gdp
```

```python
cols_to_drop = []
for y in range(1970, 1986):
    cols_to_drop.append(str(y))
for z in range(2020, 2101, 5):
    cols_to_drop.append(str(z))
cols_to_drop.append('Unnamed: 69')
```

```python
usa_per_gdp = usa_per_gdp.drop(cols_to_drop, axis=1)
usa_per_gdp
```

```python
x = list(range(1986, 2016, 5))
cols_to_select = list(range(1986, 2015, 1))
cols_to_select
y = usa_per_gdp[cols_to_select]
y
```

```python
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(8, 8), dpi=100)

ax.plot(x, y, color='m', alpha=0.6 label='GDP')
```
